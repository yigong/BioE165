```python
import numpy as np
%pylab
```

```python
n = 100
t = np.linspace(0,10e-3,n)
G0 = 40e-3
Tmax = 10e-3
N = 32
gamma_bar = 43e6
Gx = G0*t/Tmax*np.cos(2*np.pi*N*t/Tmax)
kx = gamma_bar*np.cumsum(Gx)/n
Gy = G0*t/Tmax*np.sin(2*np.pi*N*t/Tmax)
ky = gamma_bar*np.cumsum(Gy)/n
```

```python
plt.plot(kx, ky, 'o')
xlabel('$kx[m^{-1}]$')
ylabel('$ky[m^{-1}]$')
```
